# Tic Tac Toe

This is a program that will play tic tac toe

In order to play, run all of the cells, and then type **game_start()** in an empty cell.

The outline of the program is as follows:

    1. Game start function
        a. Print Rules
        b. Ask for the two players names'
        c. Tells the players if they will be X or O
        d. Prints the empty board
        e. Go to function for player1's choice
    2. player1's move function
        a. Gives player a choices for where he can place an X
        b. Asks player where he would like to place his X
        c. If X is not placed at one of the given choices, ask player once again where he would like to place his X
        d. If X is placed at one of the given choices, update the board to have an X at that choice
        e. Take away that choice so it cannot be used again
        f. check if win, if not check if board is full, if not go to player2 move
        g. if board is full, it is a tie
        h. if win, player 1 wins.
    3. player2's move function
        a. same thing as player1's move function except with player 2 stuff

```python
from time import sleep

def game_start():
    
    """Begins the game of Tic Tac Toe.
    
    Inputs players names as global variable "player1" and "player2", and prints out the board."""
    
    #variables that need to be made global
    global player1
    global player2
    global board_top
    global board_middle
    global board_bottom
    global choices
    global Win
    
    #Win is a boolean that if True, the game is over and someone has won
    Win = False
    
    print("Welcome! Let's play Tic Tac Toe! \n")
    sleep(1)
    player1 = input("Player 1, what is your name? ")
    sleep(1)
    print("Okay %s, you will be X \n" %(player1))
    sleep(1)
    player2 = input("Player 2, what is your name? ")
    sleep(1)
    print("Okay %s, you will be O \n" %(player2))
    sleep(1)
    board_top = ["-", "-", "-"]
    board_middle = ["-", "-", "-"]
    board_bottom = ["-", "-", "-"]
    
    print("Here is what the board looks like: \n {x} \n {y} \n {z}".format(x = board_top, y = board_middle, z = board_bottom))
    sleep(1)
    
    #possible areas that one could put an X or an O
    choices = ["top left", "top middle", "top right", "middle left", "middle middle", "middle right", "bottom left", "bottom middle", "bottom right"]
    
    #Go to player1's move
    player1_move()
```

```python
def player1_move():
    
    global player1
    global player2
    global board_top
    global board_middle
    global board_bottom
    global choices
    global Win
    
    #Choose where to place your X
    placement = str(input("Okay %s, where would you like to place your X? \n \nYour choices are: \n%s" %(player1, choices)))
    sleep(1)
    
    #If an available placement is not chosen then ask again
    while placement not in choices:
        print("That is not a legal move. Please choose again \n")
        sleep(1)
        placement = str(input("Where would you like to place your X? \n \nYour choices are: \n%s" %(choices)))
        sleep(1)
    
    #Placing Xs on the board
    if placement == "top left":
        board_top[0] = "X"
    elif placement == "top middle":
        board_top[1] = "X"
    elif placement == "top right":
        board_top[2] = "X"
    elif placement == "middle left":
        board_middle[0] = "X"
    elif placement == "middle middle":
        board_middle[1] = "X"
    elif placement == "middle right":
        board_middle[2] = "X"
    elif placement == "bottom left":
        board_bottom[0] = "X"
    elif placement == "bottom middle":
        board_bottom[1] = "X"
    elif placement == "bottom right":
        board_bottom[2] = "X"
    
    #Show what the board looks like with placement
    print("Here is what the board looks like: \n {x} \n {y} \n {z}".format(x = board_top, y = board_middle, z = board_bottom))
    sleep(1)
    
    #Remove the placement from the possible choices
    choices.remove(placement)
    
    #Check if player1 has won
    if board_top == ["X", "X", "X"]: 
        Win = True
    elif board_middle == ["X", "X", "X"]: 
        Win = True
    elif board_bottom == ["X", "X", "X"]:
        Win = True
    elif [board_top[0], board_middle[0], board_bottom[0]] == ["X", "X", "X"]:
        Win = True
    elif [board_top[1], board_middle[1], board_bottom[1]] == ["X", "X", "X"]:
        Win = True
    elif [board_top[2], board_middle[2], board_bottom[2]] == ["X", "X", "X"]:
        Win = True
    elif [board_top[0], board_middle[1], board_bottom[2]] == ["X", "X", "X"]:
        Win = True
    elif [board_top[2], board_middle[1], board_bottom[0]] == ["X", "X", "X"]:
        Win = True
    
    
    #Statements to direct where to go next
    if Win == True:
        print("%s wins!" %(player1))
    elif "-" not in board_top and "-" not in board_middle and "-" not in board_bottom:
        print("No winner! It's a tie!")
    else:
        player2_move()
   
```
